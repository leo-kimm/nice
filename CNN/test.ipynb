```python
from tensorflow import keras
from tensorflow.keras.models import load_model
# from keras.layers import Activation

mnist_model = load_model('model.h5')
```

```python
import cv2

image_4 = cv2.imread('drawisland.png', cv2.IMREAD_GRAYSCALE)
print(image_4.shape)
```

```python
image_4_re = cv2.resize(image_4, dsize=(28,28))
print(image_4_re.shape)
```

```python
result = mnist_model.predict(image_4_re.reshape(1,28,28,1))
print(result)
```

```python
import matplotlib.pyplot as plt
plt.imshow(image_4)
```

```python
image_size = (28,28)
image_shape = (1,28,28,1)
image_names = ['drawisland (8).png','00.png','drawisland (6).png','drawisland (7).png','100.png']

for image_name in image_names:
    image = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE)
    
    image_resized = cv2.resize(image, dsize=image_size, interpolation=cv2.INTER_AREA)

    plt.imshow(image_resized, cmap= plt.cm.gray)
    plt.show()
    
    result = mnist_model.predict(image_resized.reshape(image_shape))
    print(result[0])
```
